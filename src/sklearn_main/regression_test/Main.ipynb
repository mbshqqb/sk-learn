```python
import matplotlib.pyplot as plt
import sklearn.linear_model as linear_model
```

```python
x=[[0], [1], [2]]
y=[0,1,2]
regression=linear_model.LinearRegression()
regression.fit(x, y)
```

```python
z=regression.predict(x)
plt.scatter(x,y)
plt.plot(x,z)
plt.show()
```
